```python
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, GRU, RNN, BatchNormalization, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
df3 = df3.rename(columns = {'Essays' : 'Essay'})
# Load your data
df = pd.read_csv("students.csv")

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
import pandas as pd
df1 = pd.read_csv('students1.csv')
dfJoin = pd.concat([df,df1, df2,df3,df4,df5,df6,df7, df8], ignore_index=True)
```

```python
def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [token for token in tokens if token not in string.punctuation and token not in stopwords.words('english')]
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(tokens)

dfJoin['Essay_cleaned'] = dfJoin['Essay'].apply(preprocess_text)
dfJoin['Feedback_cleaned'] = dfJoin['Feedback'].apply(preprocess_text)
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(dfJoin['Essay_cleaned'].tolist() + dfJoin['Feedback_cleaned'].tolist())

# Convert text to sequences
X = tokenizer.texts_to_sequences(dfJoin['Essay_cleaned'])
Y = tokenizer.texts_to_sequences(dfJoin['Feedback_cleaned'])

# Find the maximum length of sequences
max_input_length = max(len(seq) for seq in X)
max_output_length = max(len(seq) for seq in Y)

# Pad sequences to the same length
X = pad_sequences(X, maxlen=625, padding='post')
Y = pad_sequences(Y, maxlen=625, padding='post')

# Define vocabulary size
vocab_size = len(tokenizer.word_index) + 1
```

```python
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=65, input_length=max_input_length),
    LSTM(64, return_sequences=True),
    LSTM(64, return_sequences=True),
    BatchNormalization(),
    GRU(256, return_sequences=True),
  #  SimpleRNN(128),
    Dropout(0.1),
    Dense(64, activation='relu'),
    Dense(vocab_size, activation='softmax')  # Softmax activation for generating sequences
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Summary of the model
model.summary()
```

```python
model.fit(X, Y, epochs=30, batch_size=32, validation_split=0.3, shuffle = True)
```
